```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# TODO:Removed the negative publishing year issue
df = df[df['Publishing Year'] > 1900]
```

```python
# TODO:Checking for the null values
df.isna().sum()
df.dropna(subset = 'Book Name',inplace = True)
```

```python
# Checking for the duplication in data
df.duplicated().sum()
```

```python
df.nunique()
```

```python
# Ditribution of publishing year
import matplotlib.pyplot as plt
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Ditribution of publishing year')
plt.show()
```

```python
# Making the bar chart of no of books in each General
df['genre'].value_counts().plot(kind = 'bar')
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.title('Number of Books in Each Genre')
plt.show()
```

```python
# Groupby
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending = False)
```

```python
# Boxplot of each General
sns.boxplot(x = 'genre',y = 'Book_ratings_count',data = df)
plt.xlabel('Genre')
plt.ylabel('Book Rating Count')
plt.title('Box plot of book ratings count for each genre')
plt.show()
```

```python
# TODO:Scatter Plot of Scatter Plot Of Sales Price Vs Units Sold
plt.scatter(df['sale price'],df['units sold'])
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title('Scatter Plot Of Sales Price Vs Units Sold')
plt.show()
```

```python
# TODO: Pie chart of language distribution of Books
language_count = df['language_code'].value_counts().sort_values(ascending = True)
plt.pie(language_count,labels = language_count.index,startangle = 90,autopct = "%1.1f%%")
plt.title('Lagnguage distribution of books')
plt.show()
```

```python
# Total revenue of each publishers
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending = False)
```

```python
# Average of book rating count
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending = False).head(1)
```

```python
# Groupby on language code and counting each number of books
df.groupby('language_code').size().sort_values(ascending = False)
```

```python
# Average rating count of each author rating
df.groupby('Author_Rating')['Book_ratings_count'].var()
```

```python
# Scatter plot of book avg rating and rating count
plt.scatter(df['Book_average_rating'],df['Book_ratings_count'])
plt.xlabel('Book Average Rating')
plt.ylabel('Book Rating Count')
plt.title('Scatter plot of Bbook Avg Rating And Rating Count')
```

```python
# Bar chart of total gross sales
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum()
total_gross_sales_by_author.sort_values(ascending = False).head(20).plot(kind = 'bar')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total Gross Sales For Each Author')
plt.show()
```

```python
# Box plot of unit sold for each auto rating
sns.boxplot(x = 'Author_Rating',y = 'units sold',data = df)
plt.xlabel('Author Ratings')
plt.ylabel('Units Sold')
plt.title('Box plots of Units Sold for Each Author Ratings')
plt.show()
```

```python
# Line plot of Total Units Sold Over The Years
df.groupby('Publishing Year')['units sold'].sum().plot(kind = 'line',marker = 'o')
plt.xlabel('Publishing Year')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold Over The Years')
plt.show()
```
